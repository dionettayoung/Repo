# Determining the best portfolio allocation
- function created to generate randomnised portfolio weights based on the same variety of stocks in the basket

```python
def create_port(log_returns, no_port, trading_days):
    # returns = dataframe of logarithmic returns for indiv stocks
    # no_port = number of portfolios to be generated
    
    # lists to store respective values for each portfolio
    weight = []
    ret = []
    var = []
    stdev = []
    
    for i in range(no_port):
        # determine indiv stock weights, normalised
        indiv_weight = np.random.random(len(log_returns.columns))
        indiv_weight /= np.sum(indiv_weight)
        weight.append(indiv_weight)
        
        # calculate values for each portfolio
        ret.append(np.sum(log_returns.mean() * indiv_weight) * trading_days)
                
        indiv_var = np.dot(indiv_weight.T, np.dot(log_returns.cov() * trading_days, indiv_weight))
        var.append(indiv_var)
        
        stdev.append(np.sqrt(indiv_var))
        
    return weight, ret, var, stdev
```

```python
sim_weight, sim_ret, sim_var, sim_stdev = create_port(log_ret, 5000, 252)

print("weights")
print(np.array(sim_weight))
print("\n")
print("returns")
print(np.array(sim_ret))
print("\n")
print("variance")
print(np.array(sim_var))
print("\n")
print("stdev")
print(np.array(sim_stdev))
```

```python
plt.figure(figsize=(15, 15))
risk_free = 0.01178
sim_sharpe = (np.array(sim_ret) - risk_free) / np.array(sim_stdev)
plt.scatter(np.array(sim_stdev), np.array(sim_ret), c=sim_sharpe, marker='o')
plt.title('Random Generated Portfolios')
plt.xlabel("Expected stdev (ie. volatility)")
plt.ylabel("Expected Returns")
plt.grid(True)
plt.colorbar(label="Sharpe Ratio")
```

# Efficient Frontier
### Upper Bound: Maximise Sharpe Ratio

```python
import scipy.optimize as opt
```

scipy does optimisation in the form minimization, hence we flip the sharpe ratio here into its negative values.


ie. find the minimum value out of all sharpe ratio values

```python
# assume equal distribution of weights across all stocks in the basket to begin with
init_weight = np.full(len(stock_ticker), 1.0 / len(stock_ticker))

def get_ret_vol_sr(weight):
    weight = np.array(weight)
    ret = np.sum(log_ret.mean() * weight) * 252
    vol = np.sqrt(np.dot(weight.T, np.dot(log_ret.cov() * 252, weight)))
    sr = (ret - risk_free) / vol
    return np.array([ret, vol, sr])

def neg_sharpe(weight):
# the number 2 is the sharpe ratio index from the get_ret_vol_sr
    return get_ret_vol_sr(weight)[2] * (-1)

def check_sum(weight):
    # return 0 if sum of the weights is 1
    return np.sum(weight) - 1

cons = ({'type': 'eq', 'fun': check_sum})
bnds = tuple((0, 1) for x in range(len(stock_ticker)))

max_result = opt.minimize(neg_sharpe, init_weight, method='SLSQP', bounds=bnds, constraints=cons)
max_result
```

### Calculate Efficient Frontier for all values

```python
frontier_y = np.linspace(0, 0.4, 300)
```

```python
def min_volatility(weight):
    return get_ret_vol_sr(weight)[1]
```

```python
frontier_x = []

for possible_return in frontier_y:
    cons = ({'type':'eq', 'fun':check_sum},
            {'type':'eq', 'fun': lambda w: get_ret_vol_sr(w)[0] - possible_return})
    
    result = opt.minimize(min_volatility,init_weight,method='SLSQP', bounds=bnds, constraints=cons)
    frontier_x.append(result['fun'])
```

```python
plt.figure(figsize=(12,8))
# plt.scatter(vol_arr, ret_arr, c=sharpe_arr, cmap='viridis')
plt.scatter(np.array(sim_stdev), np.array(sim_ret), c=sim_sharpe, marker='o')
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Volatility')
plt.ylabel('Return')
plt.plot(frontier_x,frontier_y, 'r--', linewidth=3)
plt.scatter(sim_stdev[idx], sim_ret[idx], c='red', marker='D', s=200)
plt.scatter(port_stats[1], port_stats[0], c='blue', marker='*', s=400)
plt.scatter(port_min_stats[1], port_min_stats[0], c='blue', marker='*', s=400)
plt.title('Efficient Frontier Using 15 US Stocks')
plt.show()
```

The efficient frontier visualises the optimal portfolio values that maximises expected returns got each level of expected volatility (ie. standard deviation). The blue stars represents the minimum variance portfolio (left) and maximum Sharpe ratio portfolio (right).

The red diamond represents the portfolio with the maximum Sharpe ratio out of all the randomly generated portfolios earlier above. As seen from the plot, none of the simulated portfolios lie on the efficient frontier, and this is because the simulated portfolios provide a non-zero positive weight to all stocks in the basket. However, the portfolios on the efficient frontier do not necessarily do so. Some stocks were given a weight of zero, such that the entire portfolio's return will be maximised for a given level of volatility.

Therefore, the simulated portfolios are not aligned with the efficient frontier.

# Capital Market Line


The current visualisation of the efficient frontier only consists of individual points. To be a proper efficient frontier we would like it to be continuous line. We can interpolate a B-spline representation of the data points to create a continuous line and give us a continuous function of the efficient frontier.

```python
# Import cubic splines interpolation module
import scipy.interpolate as sci

# Select the index of the absolute minimum variance portfolio
min_var = np.argmin(frontier_x)

# Select expected volatilities for the stocks on the efficient frontier
ex_stds = frontier_x[min_var:]

# Select expected returns for the stocks on the efficient frontier
ex_rs = frontier_y[min_var:]

# Interpolate the B-spline representation of the data points
tck = sci.splrep(ex_stds, ex_rs)
```

```python
# Obtain coordinates to be used for plotting the B-spline interpolation of the efficient frontier
x_lower = min(ex_stds)
x_upper = max(ex_stds)
x_range = np.linspace(x_lower, x_upper, 500)
bspline = sci.splev(x_range, tck)
```

```python
plt.figure(figsize=(12,8))
plt.scatter(np.array(sim_stdev), np.array(sim_ret), c=sim_sharpe, marker='o')
plt.plot(x_range, bspline)
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Volatility')
plt.ylabel('Return')
#plt.plot(frontier_x,frontier_y, 'r--', linewidth=3)
plt.scatter(sim_stdev[idx], sim_ret[idx], c='red', marker='D', s=200)
plt.scatter(port_stats[1], port_stats[0], c='blue', marker='*', s=400)
plt.scatter(port_min_stats[1], port_min_stats[0], c='blue', marker='*', s=400)
plt.title('Efficient Frontier Using 15 US Stocks')
plt.show()
```

The B-spline interpolation can be used to find a continuously differentiable function and its derivatives

```python
# Define a continuously differentiable function f(x) for the efficient frontier
def f(x):
    return sci.splev(x, tck, der=0)

# Define te first derivate function of f(x)
def df(x):
    return sci.splev(x, tck, der=1)
```

Assuming a risk-free rate of return of 1.178% and a parameter set p = (a,b,x) we can set up a system of equations for the capital market line.

```python
# Set up a system of equations for the conditions that the CML has to satisfy
def cml_conditions(p, rf = risk_free):
    eq1 = rf - p[0]
    eq2 = rf + p[1] * p[2] - f(p[2])
    eq3 = p[1] - df(p[2])
    return eq1, eq2, eq3

# Solve for the parameters of the Capital Market Line
cml = opt.fsolve(cml_conditions, [0.01178, 0.11, 0.13]) # initial parameterisation

# Create input variable for CML
cx = np.linspace(0.17, 0.3)
```

```python
# Plot the resulting Capital Market Line together with the efficient frontier

plt.figure(figsize=(20,15))
plt.scatter(np.array(sim_stdev), np.array(sim_ret), c=sim_sharpe, marker='o')
plt.plot(x_range, bspline, label = "Efficient Frontier")
plt.plot(cx, cml[0] + cml[1] * cx, lw=1.5, label='Capital Market Line')
plt.plot(cml[2], f(cml[2]), 'b*', markersize=15.0)
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Volatility')
plt.ylabel('Return')
plt.grid(True)
plt.scatter(sim_stdev[idx], sim_ret[idx], c='red', marker='D', s=200)
plt.scatter(port_stats[1], port_stats[0], c='blue', marker='*', s=400)
plt.scatter(port_min_stats[1], port_min_stats[0], c='blue', marker='*', s=400)
plt.title('Efficient Frontier and Capital Market Line For 15 US Stocks')
plt.legend(loc=5)
plt.xlim(0.17, 0.3)
plt.show()
```

In the visualisation above, the upper blue star represents the optimal portfolio from the efficient frontier where the tangent line passes through the riskless asset.

```python
# Obtain coordinates to be used for plotting the B-spline interpolation of the efficient frontier
cx_lower = min(cex_stds)
cx_upper = max(cex_stds)
cx_range = np.linspace(cx_lower, cx_upper, 500)
cbspline = sci.splev(cx_range, ctck)
```

```python
# Define a continuously differentiable function f(x) for the efficient frontier
def comb_f(x):
    return sci.splev(x, ctck, der=0)

# Define te first derivate function of f(x)
def comb_df(x):
    return sci.splev(x, ctck, der=1)
```

```python
# Set up a system of equations for the conditions that the CML has to satisfy
def comb_cml_conditions(p, rf = risk_free):
    eq1 = rf - p[0]
    eq2 = rf + p[1] * p[2] - comb_f(p[2])
    eq3 = p[1] - comb_df(p[2])
    return eq1, eq2, eq3

# Solve for the parameters of the Capital Market Line
comb_cml = opt.fsolve(comb_cml_conditions, [0.01178, 0.11, 0.20]) # initial parameterisation

# Create input variable for CML
comb_cx = np.linspace(0.17, 0.3)
```

```python
# Plot the resulting Capital Market Line together with the efficient frontier

plt.figure(figsize=(12,8))
plt.scatter(np.array(csim_stdev), np.array(csim_ret), c=csim_sharpe, marker='o')
plt.plot(comb_cx, comb_cml[0] + comb_cml[1] * comb_cx, lw=1.5, label='Capital Market Line')
plt.plot(cx_range, cbspline)
plt.colorbar(label='Sharpe Ratio')
plt.xlabel('Volatility')
plt.ylabel('Return')
plt.scatter(csim_stdev[cidx], csim_ret[cidx], c='red', marker='D', s=200)
plt.scatter(comb_port_stats[1], comb_port_stats[0], c='blue', marker='*', s=400)
plt.scatter(comb_port_min_stats[1], comb_port_min_stats[0], c='blue', marker='*', s=400)
plt.title('Efficient Frontier Using US Stocks and Cryptocurrency')
plt.show()
```
